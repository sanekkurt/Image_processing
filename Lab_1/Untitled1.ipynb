```python
def sravnenie_kartinok(colvo_pix, s_chem_sravnivat, chto_sravnivat):
    dic = {}
    for i in range(colvo_pix):
        imgRow = np.roll(chto_sravnivat, i, 0)
        for j in range(colvo_pix):
            imgRowCol = np.roll(imgRow, j, 1)
            correlation = (imgRowCol * s_chem_sravnivat).sum()
            pix = [i,j]
            dic.update({correlation: pix})
    a = list(dic.keys())
    maxcorrelation = max(a)
    shiftRow = dic.get(maxcorrelation)[0]
    shiftCol = dic.get(maxcorrelation)[1]
    temp = np.roll(chto_sravnivat, shiftRow, 0)
    finalPict = np.roll(temp, shiftCol, 1)
    return finalPict
```

```python
b_1 = sravnenie_kartinok(5, g1, b1)
r_1 = sravnenie_kartinok(5, g1, r1)
b_2 = sravnenie_kartinok(5, g2, b2)
r_2 = sravnenie_kartinok(5, g2, r2)
b_3 = sravnenie_kartinok(5, g3, b3)
r_3 = sravnenie_kartinok(5, g3, r3)
```

```python
img_comb1 = np.dstack((b_1,g1,r_1))
imshow(img_comb1)
imsave('1.jpg', img_comb1)
```

```python
img_comb2 = np.dstack((b_2,g2,r_2))
imshow(img_comb2)
imsave('2.jpg', img_comb2)
```

```python
img_comb3 = np.dstack((b_3,g3,r_3))
imshow(img_comb3)
imsave('3.jpg', img_comb3)
```
